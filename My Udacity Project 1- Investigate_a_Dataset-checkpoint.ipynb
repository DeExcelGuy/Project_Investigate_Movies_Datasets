### Question 1: Do movies with higher budget generate more box office revenue?<a id='Q1'></a>

##### Let's see the correlation between Budget and Revenue using the pearson method

```python
df[['budget','revenue']].corr(method='pearson')
```

> We can see there is a high correlation between revenue and budget, which is about 69%. That is to say the higher the budget the higher the revenue generated.

##### Let's dive deep in by plotting a line chart to show the trend year on year 👇

```python
# Let's group the revenue and budget by year

sb.set_palette("Set2",50,.80) # this is just to influence the chart colour
df.groupby(['release_year']).revenue.sum().plot(kind='line',figsize = (10,6)) #revenue line
df.groupby(['release_year']).budget.sum().plot(kind='line',color='r',figsize = (10,6)) #budget line

# Let's plot the revenue and budget year on year

plt.xlabel("release_Yr")
plt.ylabel("budget/revenue")
plt.title("budget/revenue trend")
plt.legend(loc=0);
```

> We can see that the revenue grows as the budget grows which supports the strong correlation of 69%.

##### Let us check the correlation between the revenue, budget and popularity

```python
cor_df = df[['revenue', 'budget', 'popularity']]

sb.set(rc={'figure.figsize':(25,25)}, font_scale=1.3, style="ticks")
sb.set_palette("Set2",50,.80) # this is just to influence the chart colour

chart0 = sb.jointplot(x = "budget", y = "revenue", kind = "scatter", data = cor_df)
chart0.fig.suptitle('relation: budget & revenue')

chart1 = sb.jointplot(x = "revenue", y = "popularity", kind = "scatter", data = cor_df)
chart1.fig.suptitle('relation: revenue & popularity')

chart2 = sb.jointplot(x = "budget", y = "popularity", kind = "scatter", data = cor_df)
chart2.fig.suptitle('relation: budget & popularity');
```

> There is a positive relation between the budget, revenue and popularity.
>
> We can see revenue increases as popularity increases with 61% correlation. Same with budget and popularity.

##### Let's check the distribution using histogram

```python
# Let's print the maximum budget and revenue
max_budget =df["budget"].max()
max_revenue =df["revenue"].max()

# Let's also print the mean budget and revenue
mean_budget =df["budget"].mean()
mean_revenue =df["revenue"].mean()

# Let's also print the mean budget and revenue
median_budget =df["budget"].median()
median_revenue =df["revenue"].median()

print("max_budget={:,}\nmax_revenue={:,}\n\nmean_budget={:,.2f}\nmean_revenue={:,.2f}\n\nmedian_budget={:,.2f}"
      "\nmedian_revenue={:,.2f}".format(max_budget,max_revenue,mean_budget,mean_revenue,median_budget,median_revenue))

# Let's plot two histograms of the revenue and budget

sb.set_palette("Set2",50,.80) # this is just to influence the chart colour
df[["budget","revenue"]].hist(figsize=(15,5))
plt.legend();
```

### Question 2:  Which movie genres are most popular over time?<a id='Q2'></a>

##### Let's calculate the number of time a genre is mentioned

```python
# Let split the genre into string and the count each genre using a function

def separate(x):

    step1= df[x].str.cat(sep = '|')
    separate = pd.Series(step1.split('|'))
    step2 = separate.value_counts(ascending=False)
    return step2
```

##### Let's display the result

```python
count_of_genre = separate('genres')
print(count_of_genre);
```

> We now have the count of genre

##### Let's plot the genre on a bar chart

```python
# let's sort the count_of_genre in ascending order by value
count_of_genre = count_of_genre.sort_values(ascending=True)

# let's plot the bar chart
sb.set_palette("Set2",50,.80) ### this is just to influence the chart colour
count_of_genre.plot(kind= 'barh',figsize = (13,6))

# let's add the axis labels
plt.title('highest genre of movie')
plt.xlabel('#_of_movies')
plt.ylabel('genre');
```

> We can see that drama is the most produced genre in the movie industry.

##### Let's count the number of genre per year

```python
# let's define the set for the year and the genre

yr_set = set()
genre_set = set()
genres_release_yr = df[['genres', 'release_year']]

# let's get the unique release_year

unique_release_yrs = genres_release_yr['release_year'].drop_duplicates()

for year in unique_release_yrs:
    if year not in yr_set:
        yr_set.add(year)
        
# let's get the pair genres/release_year for all the years

for year in yr_set:
    genre_dict = {}
    genres_in_year = genres_release_yr[genres_release_yr.release_year == year]
    genres_in_year = genres_in_year['genres'].values
    for item in genres_in_year:
        genres_row = item.split('|')
        for genre in genres_row:
            if genre not in genre_set:
                genre_set.add(genre)

# let's create a dataframe

genre_count_per_year_df = pd.DataFrame(index = yr_set, columns=genre_set)
genre_count_per_year_df[:] = 0  

for year in yr_set:
    genre_dict = {}
    genres_in_year = genres_release_yr[genres_release_yr.release_year == year]
    genres_in_year = genres_in_year['genres'].values
    for item in genres_in_year:
        genres_row = item.split('|')
        for genre in genres_row:
            if genre not in genre_dict:
                genre_dict[genre] = 1
            else:
                genre_dict[genre] +=1
                    
    aux_df = pd.DataFrame(genre_dict, index = [year])
    genre_count_per_year_df.loc[year, aux_df.columns] = genre_count_per_year_df.loc[year, aux_df.columns] + aux_df.loc[year]

#count of genre of movies per year

count_of_genre_by_year = pd.DataFrame([genre_count_per_year_df.idxmax(axis = 1).values,
                                          genre_count_per_year_df.apply( max, axis=1 ).values],
                                          columns = genre_count_per_year_df.index,
                                         index = ['genre', 'counts'])

count_of_genre_by_year = count_of_genre_by_year.T
```

##### Let's display the result

```python
# let's display the first 15 genres

count_of_genre_by_year.head(15)
```

##### Let's plot the above data 👆 

```python
# let's set the seaborn setting
sb.set(rc={'figure.figsize':(15,13)}, font_scale=1)
sb.set_palette("Set2",50,.80)

# plotting using stacked area chart
chart0 = genre_count_per_year_df.plot.area(stacked=True);
chart0.set(xlabel='years', ylabel='#_of_movies', title = 'stacked area chart: Trend of movie genres')
plt.show()
```

> We can say that the movie industry has really grown in the last 3 decades. The Drama genre is the most produced over time.
